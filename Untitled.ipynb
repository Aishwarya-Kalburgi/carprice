```python
import pandas as pd
```

```python
df=pd.read_csv("car data.csv")
```

```python
df.head()
```

```python
print(df["Fuel_Type"].unique())
print(df["Seller_Type"].unique())
print(df["Transmission"].unique())
print(df["Owner"].unique())
```

```python
df.columns
```

```python
dataset = df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
```

```python
dataset.head()
```

```python
dataset['Curr_Yr']=2020
```

```python
dataset.head()
```

```python
dataset['No.Yrs']=dataset['Curr_Yr']-dataset['Year']
```

```python
dataset.head()
```

```python
dataset.drop(['Year'],axis=1,inplace=True)
```

```python
dataset.drop(['Curr_Yr'],axis=1,inplace=True)
```

```python
dataset.head()
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset.head()
```

```python
dataset.corr()
```

```python
import seaborn as sns
```

```python
sns.pairplot(dataset)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
corr=dataset.corr()
top_corr_features = corr.index
plt.figure(figsize=(20,20))
p=sns.heatmap(dataset[top_corr_features].corr(),annot=True,cmap='RdYlGn')
```

```python
x=dataset.iloc[:,1:]
y=dataset.iloc[:,0]
```

```python
y.head()
```

```python
from sklearn.ensemble import ExtraTreesRegressor
model=ExtraTreesRegressor()
model.fit(x,y)
```

```python
print(model.feature_importances_)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
from sklearn.ensemble import RandomForestRegressor
rfg=RandomForestRegressor()
```

```python
import numpy as np
n_estimators=[int(x) for x in np.linspace(start=100, stop=1200, num=12)]
max_depth=[int(x) for x in np.linspace(start=5, stop=30, num=6)]
max_features=['auto','sqrt']
min_samples_split = [2,5,10,15,100]
min_samples_leaf =[1,2,5,10]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_grid={'n_estimators':n_estimators,
             'max_depth':max_depth,
             'max_features':max_features,
             'min_samples_split':min_samples_split,
             'min_samples_leaf':min_samples_leaf}
```

```python
print(random_grid)
```

```python
rf_random=RandomizedSearchCV(rfg, param_distributions=random_grid,scoring='neg_mean_squared_error', n_iter=10, cv=5, verbose=2, random_state=42, n_jobs=1)
```

```python
rf_random.fit(x_train,y_train)
```

```python
predictions=rf_random.predict(x_test)
```

```python
predictions
```

```python
sns.distplot(y_test-predictions)
```

```python
plt.scatter(y_test,predictions)
```

```python
import pickle 
file = open('random_forest_regression_model.pkl','wb')
pickle.dump(rf_random,file)
```
